## Задание 1
**Даны 2 переменных, в которых хранятся строки произвольной длины: phrase_1 и phrase_2.
Напишите код, который проверяет какая из этих строк длиннее.**

```python
# Определяем переменные, ввод пользователя
phrase_1 = input('Введите первую фразу: ')
phrase_2 = input('Введите вторую фразу: ')

if len(phrase_1) > len(phrase_2):
    print('Первая фраза длиннее второй фразы')

elif len(phrase_1) == len(phrase_2):
    print('Фразы равной длины')

else:
    print('Вторая фраза длиннее первой фразы')
```

## Задание 2
**Дана переменная, в которой хранится четырехзначное число (год). Необходимо написать программу, которая выведет, является ли данный год високосным или обычным.**

```python
# Определяем переменную, воод пользователя
year = int(input('Введите год: '))

# Первая проверка - если год не делится на 4 без остатка, то он не високосный
if (year % 4 != 0):
    print('Обычный год')
    
# Столетия, делящиеся на 4 без остатка, являются исключениями (могут быть и не високосными)
# Следовательно, надо проверить, не является ли год, делящийся без остатка, столетием
# Если этот год - не столетие, то он високосный
elif (year % 100 != 0):
    print('Високосный год')

# Если год является столетием, то его надо разделить на 400 (100*4)
# Если это деление без остатка, то год високосный
elif (year % 400 == 0):
    print('Високосный год')
    
# Все оставшиеся столетия - не високосные
else:
    print('Обычный год')
```

- **если каждое из трех измерений менее 15 сантиметров, то выведите на экран "Коробка №1";**

## Задание 5 (необязательное)

**Дана переменная, в которой хранится шестизначное число (номер проездного билета). Напишите программу, которая будет определять, является ли данный билет “счастливым”. Билет считается счастливым, если сумма первых трех цифр совпадает с суммой последних трех цифр номера.**

```python
# Определяем переменную, где будет храниться номер билета, пользовательский ввод
ticket = input('Введите номер билета: ')

# Проверяем, чтобы пользователь не вносил билет с неправильным количеством цифр
if len(ticket) != 6:
    print('Неправильный номер билета!')
else:
    # Преобразуем строку, занесенную пользователем, в список символов
    ticket = list(ticket)
    
    # Преобразуем каждый элемент списка в целое число
    ticket = list(map(int, ticket))

    # Суммируем первые три и последние три элемента списка и сравниваем их значение, выводя результат на экран
    if sum(ticket[0:3]) == sum(ticket[3:6]):
        print('Поздравляем, у вас счастливый билет!')
    else:
        print('Это обычный билет.')
```

## Задание 6 (необязательное)
**Напишите программу, которая сможет вычислять площади трех фигур (круг, треугольник и прямоугольник). Тип фигуры запрашиваем через пользовательский ввод, после чего делаем запрос характеристик фигуры:**

- **если пользователь выбрал круг, запрашиваем его радиус,**
- **если треугольник – длины трех его сторон;**
- **если прямоугольник – длины двух его сторон.**

```python
# Импортируем библиотеку math, чтобы использовать оттуда число Пи и функцию квадратного корня
import math
shape = input('Введите тип фигуры (круг, треугольник или прямоугольник): ')

# Чтобы унифицировать написание формы фигуры, все буквы сделаем прописными
shape = shape.upper()

if shape == 'КРУГ':
    radius = int(input('Введите радиус круга: '))
    area = math.pi*(radius**2)
    print('Площадь круга = ', area)
    
elif shape == 'ТРЕУГОЛЬНИК':
    side_a = int(input('Введите длину первой стороны треугольника: '))
    side_b = int(input('Введите длину второй стороны треугольника: '))
    side_c = int(input('Введите длину третей стороны треугольника: '))
    
    # Проверяем, может ли фигура с такой длиной сторон быть треугольником и, если может:
    if (side_a + side_b > side_c) and (side_a + side_c > side_b) and (side_b + side_c > side_a):
    
        # Вычисляем половину периметра
        h_perim = (side_a + side_b + side_c)/2
    
        # Вычисляем площадь треугольника, используя формулу Герона
        area = math.sqrt(h_perim * (h_perim - side_a) * (h_perim - side_b) * (h_perim - side_c))
        print('Площадь треугольника = ', area)
    
    # Если фигура не может быть треугольником
    else:
        print('Треугольника с такими сторонами не может существовать!')
        
elif shape == 'ПРЯМОУГОЛЬНИК':
    side_a = int(input('Введите длину первой стороны прямоугольника: '))
    side_b = int(input('Введите длину второй стороны прямоугольника: '))
        
    area = side_a * side_b
    print('Площадь прямоугольника равна: ', area)
    
else:
    print('Неправильный тип фигуры!')
```
